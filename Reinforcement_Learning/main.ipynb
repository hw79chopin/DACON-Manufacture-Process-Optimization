```python
from model_Roddie import DACON
```

```python
if __name__ == '__main__':
    tf.compat.v1.disable_eager_execution()
    env = DACON()
    lr = 0.001
    n_games = 100
    agent = Agent(gamma=0.99, epsilon = 0.001, lr=lr, input_dims = 15, 
                 n_actions = 18*18, mem_size = 10000, batch_size = 64, epsilon_end = 0.001)
    scores = []
    eps_history = []
    action_list = []
    for i in tqdm(range(n_games)):
        done = False
        score = 0
        observation = env.reset()
        print(i,'번째 epoch 돔')
        while not done:
            action = agent.choose_action(observation)
            
            
            action1 = action // 18
            action2 = action % 18
            
            print(action1, action2)
            observation_, reward, done  = env.step(action1, action2, observation)
            
            score += reward
            agent.store_transition(observation, action1*18+ action2, reward, observation_, done)
            observation = observation_
            agent.learn()
        
        eps_history.append(agent.epsilon)
        scores.append(score)
        print('episode :',i, 'score %.2f' % score)
```
